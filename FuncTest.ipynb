```python
from matplotlib import pyplot as plt
```

```python
import numpy as np
```

```python
def func1(x):
    y = np.sqrt(2*x - np.power(x, 3))
    return y
    
```

```python
x = np.linspace( 0, np.sqrt(2), 256, endpoint=True)
x1 = np.linspace(0, np.sqrt(2),256, endpoint=True)
X = np.concatenate((x, x1))
Y = func1(X)
```

```python
# Create a figure of size 8x6 inches, 80 dots per inch
plt.figure(figsize=(8, 6), dpi=80)

# Create a new subplot from a grid of 1x1
plt.subplot(1, 1, 1)


# Plot cosine with a blue continuous line of width 1 (pixels)
plt.plot(X, Y, color="blue", linewidth=1.0, linestyle="-", label='func1')

plt.show()
```
